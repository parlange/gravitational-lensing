```python
import numpy as np
z=np.linspace(0.0,10.0,1000)
print (z.shape, z.min(),z.max())
```

```python
from astropy.cosmology import FlatLambdaCDM
cosmo = FlatLambdaCDM(H0=70, Om0=0.3)
da=cosmo.angular_diameter_distance(z)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig,ax=plt.subplots(1,1,figsize=(8,8))
ax.plot(z,da,'-')
```

```python
z1=0.5
z2=3.0

da1=cosmo.angular_diameter_distance(z1)
da2=cosmo.angular_diameter_distance(z2)
print(da1)
print(da2)
print(da2-da1)
```

```python
da12=cosmo.angular_diameter_distance_z1z2(z1,z2)
print(da12)
```
